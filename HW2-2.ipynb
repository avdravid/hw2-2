```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation times for each particle count (in seconds)
a = 0.0660043  # 1000 particles
b = 0.0720346  # 10000 particles
c = 0.137796   # 100000 particles
d = 1.00271    # 1000000 particles
e = 6.98581    # 6000000 particles

# Convert times to ms or s where appropriate
def format_time(time):
    if time < 1e-6:
        return f"{time * 1e9:.2f} ns"
    elif time < 1e-3:
        return f"{time * 1e6:.2f} µs"
    elif time < 1:
        return f"{time * 1e3:.2f} ms"
    else:
        return f"{time:.2f} s"

# Plotting the data
plt.plot([1000, 10000, 100000, 1000000, 6000000], [a, b, c, d, e], label="MPI 2D Box")
plt.yscale('log')
plt.xscale('log')
plt.grid()

plt.xlabel("# Particles")
plt.ylabel("Time (s)")
plt.legend()

plt.title("Performance of MPI 2D Box Implementation")
plt.legend(title="Times:\n"
                f"1000 particles: {format_time(a)}\n"
                f"10000 particles: {format_time(b)}\n"
                f"100000 particles: {format_time(c)}\n"
                f"1000000 particles: {format_time(d)}\n"
                f"6000000 particles: {format_time(e)}", loc="upper left")

plt.show()
```
